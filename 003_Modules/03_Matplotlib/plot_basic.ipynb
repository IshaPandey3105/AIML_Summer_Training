## ðŸ“Œ Common Plot Types with Matplotlib
# Plot Type	        Function
# Line Plot	        plt.plot()
# Bar Chart	        plt.bar()
# Histogram	        plt.hist()
# Scatter Plot	    plt.scatter()
# Pie Chart	        plt.pie()

```python
# ðŸ”¹Line Plot
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,30,40]
plt.plot(x, y)                 # Plot line chart
plt.title("Simple Line Graph") # Add a title
plt.xlabel("X Axis")           # Label X
plt.ylabel("Y Axis")           # Label Y
plt.grid(True)                 # Show grid
plt.show()                     # Display the plot
```

```python
# ðŸ”¹Bar Chart
import matplotlib.pyplot as plt
x = ["AI", "ML", "DL"]
y = [90, 85, 80]

plt.bar(x, y, color='purple',edgecolor='yellow')
plt.title("Subject Scores")
plt.show()
```

```python
# ðŸ”¹Scatter Plot
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 20, 25]

plt.scatter(x, y, color='green')
plt.title("Scatter Example")
plt.show()
```

```python
# ðŸ”¹Pie Chart
import matplotlib.pyplot as plt
_label = ["Python", "Java", "C++"]
sizes = [60, 25, 15]
plt.pie(sizes, labels=_label, autopct="%.0f%%")
plt.title("Programming Language Usage")
plt.show()
```

```python
# ðŸ”¹Histogram
import matplotlib.pyplot as plt

# Sample data (like student marks or ages)
data = [12, 15, 17, 18, 21, 22, 22, 25, 27, 29, 30, 30, 32, 35, 35, 37, 40]

# Create histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='brown')

# Add labels and title
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```
