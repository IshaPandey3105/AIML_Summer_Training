```python
import matplotlib.pyplot as plt
import numpy as np

x = ["c", "java", "c++", "python"]
y = [10, 50, 40, 80]
y1 = [20, 30, 50, 90]

x_indexes = np.arange(len(x))  # [0, 1, 2, 3]
width = 0.4  # width of each bar

# First bar set (shift left)
plt.bar(x_indexes - width/2, y, width=width)

# Second bar set (shift right)
plt.bar(x_indexes + width/2, y1, width=width)

# Setting x-axis labels
plt.xticks(x_indexes, x)        # Replace 0,1,2,3 with language names
plt.xlabel("Languages")
plt.ylabel("Scores")
plt.title("Comparison of Two Bar Sets")                 
plt.grid(True, axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = ["c", "java", "c++", "python"]
y = [10, 50, 40, 80]
y1 = [20, 30, 50, 90]

x_indexes = np.arange(len(x))  # numeric positions for categories
width = 0.4  # bar width

# Plot both bars side by side
plt.bar(x_indexes - width/2, y, width=width, color="skyblue", edgecolor="brown", label='Set 1')
plt.bar(x_indexes + width/2, y1, width=width, color="pink", edgecolor="black", label='Set 2')

# Customizations
plt.xticks(x_indexes, x)  # Set x-axis labels to languages
# plt.xlabel("Languages")
plt.ylabel("Scores")
plt.title("Side-by-Side Bar Chart")
plt.grid(True, axis='y')
plt.legend()            #Shows which color represents "Set 1" and which is "Set 2".

plt.xlabel("______________This is a very long X-axis label that........................ might be cut off      ")

plt.tight_layout()          #Adjusts spacing to prevent label/title overlap or cutoff
plt.show() 

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["c", "java", "c++", "python"]
y = [10, 50, 40, 80]
z = [20, 30, 50, 90]

width = 0.4  # width of each bar

p = np.arange(len(x))  # [0, 1, 2, 3]
p1=[ i+width for i in p ] # [0.4, 1.4, 2.4, 3.4]  
c=['y','b','y','b']
c1=['r','m','r','m']

plt.bar(p,y,width,color=c,edgecolor="r",label="PL1",align="center",alpha=0.7,hatch='*')  # alpha = transparency
plt.bar(p1,z,width,color=c1,edgecolor="b",label="PL2",align="center",alpha=0.7,hatch='/')

# Setting x-axis labels
plt.xticks(p+width/2, x,rotation=20)        # Replace 0,1,2,3 with language names and little rotation
plt.xlabel("Languages",fontsize="10")
plt.ylabel("Scores",fontsize="10")
plt.title("Comparison of Two Bar Sets",fontsize="10")                 
plt.grid(True, axis='y')
plt.legend()
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt                                                                              
# Sample data (like student marks or ages)                                          # histtype -->	'bar', 'step', 'stepfilled'
data = [12, 15, 17, 18, 21, 22, 22, 25, 27, 29, 30, 30, 32, 35, 35, 37, 40]         # by default orientation is vertical
plt.hist(data, bins=5, color='skyblue', edgecolor='brown',cumulative=0,bottom=20,	histtype="bar"	,alpha=0.9,	rwidth=0.98,label="Marks",	orientation="vertical")#,log=True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data: Marks of students
marks = [45, 50, 52, 58, 60, 60, 61, 65, 70, 72, 75, 78, 80, 85, 90]

# Custom bins  You decide the ranges
bins = [40, 50, 60, 70, 80, 90, 100]

plt.hist(marks, bins=bins, color='m', edgecolor='black', rwidth=0.9)

plt.title('Marks Distribution Histogram')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.grid(axis='y')

plt.show()
```

```python
import matplotlib.pyplot as plt
# Sample Data: Marks of two different classes
class_A = [45, 50, 52, 58, 60, 60, 61, 65, 70, 72, 75, 78, 80, 85, 90]
class_B = [40, 48, 55, 57, 59, 62, 67, 68, 74, 76, 79, 81, 84, 88, 92]

# Same bins for both
l = [40, 50, 60, 70, 80, 90, 100]

plt.hist(class_A, bins=l, color='skyblue', edgecolor='black', rwidth=0.9, label='Class A', alpha=0.7)
plt.hist(class_B, bins=l, color='orange', edgecolor='black', rwidth=0.9, label='Class B', alpha=0.5)

plt.title('Marks Distribution: Class A vs Class B')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.legend()
plt.grid(axis='y')

plt.show()
```

###   Beautiful Side-by-Side Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

#  Sample Data (Marks)
class_A = [45, 50, 52, 58, 60, 60, 61, 65, 70, 72, 75, 78, 80, 85, 90]
class_B = [40, 48, 55, 57, 59, 62, 67, 68, 74, 76, 79, 81, 84, 88, 92]

#  Bins (Marks Ranges)
bins = [40, 50, 60, 70, 80, 90, 100]

#  Frequency Counts
counts_A, _ = np.histogram(class_A, bins)
counts_B, _ = np.histogram(class_B, bins)

#  Bar Positions
x = np.array(bins[:-1])  # Left edges of the bars
bar_width = 2

#  Plot with Beautiful Colors
plt.figure(figsize=(10, 6))  # Bigger size

plt.bar(x - bar_width/2, counts_A, width=bar_width, color='#7FB3D5', edgecolor='#154360', label='Class A')  # Soft blue
plt.bar(x + bar_width/2, counts_B, width=bar_width, color='#F5B7B1', edgecolor='#641E16', label='Class B')  # Soft pink

#  Title & Labels (Pretty Fonts)
plt.title('Marks Distribution: Class A vs Class B', fontsize=16, fontweight='bold', color="#08C024")
plt.xlabel('Marks Range', fontsize=12, color="#13EFB5")
plt.ylabel('Number of Students', fontsize=12, color="#09D2D2")

#  Legend with nicer font
plt.legend(title='Classes', fontsize=10, title_fontsize=12)

#  Soft Grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

# Pie Plot

```python
import matplotlib.pyplot as plt
y = ["Python", "Java", "C++","c"]
x = [60, 25, 15,10]
ex=[0.0,0.0,0.0,0.2]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.0f%%",shadow=True,radius=1,labeldistance=0.99,startangle=90,textprops={"fontsize":10}
        ,counterclock=True,rotatelabels=True,wedgeprops={"linewidth":3,"edgecolor":'m'})#center=(2,2))
plt.title("Programming Language Usage",fontsize=12)
plt.show()
```
