# 1ðŸ“Œ Linear Regression 

## âž¤ What is Linear Regression?
Linear Regression is a supervised learning algorithm that models the relationship between input (X) and output (y) using a straight line.

**Equation:**  
y = mX + c

- m = slope
- c = intercept

## âž¤ Use Cases:
- Predicting marks based on hours studied
- Predicting house prices
- Estimating salary from experience

## âž¤ scikit-learn Code:
```python
from sklearn.linear_model import LinearRegression

X = [[1], [2], [3], [4], [5]]
y = [20, 40, 60, 80, 100]

model = LinearRegression()
model.fit(X, y)

print("Predicted value for 6:", model.predict([[6]]))
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

## âž¤ .fit() and .predict()
- `.fit(X, y)` â†’ trains the model
- `.predict(X)` â†’ predicts output for new inputs

## âž¤ Gradient Descent (How model learns)
- Minimizes error using loss function (MSE)
- Iteratively updates m and c
- Controlled by learning rate (Î±)

## âž¤ Loss Function: Mean Squared Error (MSE)
\[
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
\]



````python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = [[1], [2], [3], [4], [5]]
y = [20, 40, 60, 80, 100]

model = LinearRegression()
model.fit(X, y)

# Plotting
plt.scatter(X, y, color='red', label='Actual Points')
plt.plot(X, model.predict(X), color='blue', label='Best Fit Line')
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.title("Linear Regression - Best Fit Line")
plt.legend()
plt.grid()
plt.show()
````

## 2ðŸ“Œ Logistic Regression â€” Full Theory

---

### ðŸ”¹ What is Logistic Regression?

- Logistic Regression is a **supervised machine learning algorithm** used for **binary classification**.
- It predicts whether a data point belongs to **Class 1 or Class 0** based on input features.
- Instead of predicting exact values (like Linear Regression), it predicts **probabilities** using the **sigmoid function**.
- Even though it has â€œregressionâ€ in its name, itâ€™s a **classification** algorithm.

---

### ðŸ”¹ Sigmoid Function

The sigmoid function is used to **squash** the output of a linear equation into a range between **0 and 1**.

    sigmoid(z) = 1 / (1 + e^(-z))

Where:
- z = w1*x1 + w2*x2 + ... + b  â†’ (linear combination)
- e is Eulerâ€™s number
- Output > 0.5 â†’ classified as Class 1
- Output < 0.5 â†’ classified as Class 0

---

### ðŸ”¹ Loss Function: Binary Cross Entropy (Log Loss)

In logistic regression, we use **Binary Cross Entropy** (also called **Log Loss**) to calculate the error.

    Loss = - [ y * log(Å·) + (1 - y) * log(1 - Å·) ]

Where:
- y = actual label (0 or 1)
- Å· = predicted probability (from sigmoid)
- Loss is high when prediction is far from actual value
- Lower loss = better model

---

## ðŸ”¹ Evaluation Metrics

When working with classification, accuracy alone is not enough (especially if the data is imbalanced). So we use these metrics:

---

### âœ… Confusion Matrix

A 2x2 table showing actual vs predicted values:

|                        | Predicted: 1 | Predicted: 0 |
|------------------------|--------------|--------------|
| Actual: 1 (Positive)   | True Positive (TP) | False Negative (FN) |
| Actual: 0 (Negative)   | False Positive (FP) | True Negative (TN) |

---

### âœ… Accuracy

    Accuracy = (TP + TN) / (TP + TN + FP + FN)

Shows the overall correctness of predictions.

---

### âœ… Precision

    Precision = TP / (TP + FP)

Of all predicted positives, how many were actually positive.

Use when **False Positives** are more costly.

---

### âœ… Recall (Sensitivity / TPR)

    Recall = TP / (TP + FN)

Of all actual positives, how many were correctly predicted.

Use when **False Negatives** are more dangerous.

---

### âœ… F1 Score

    F1 Score = 2 * (Precision * Recall) / (Precision + Recall)

- Harmonic mean of Precision and Recall.
- Use when you want a balance between both.

---

### ðŸ“ Choosing the Right Metric:

| Goal | Metric |
|------|--------|
| General performance on balanced data | Accuracy |
| Avoid false positives (e.g. spam detection) | Precision |
| Avoid false negatives (e.g. cancer test) | Recall |
| Balanced importance | F1 Score |

---


## 3ðŸŒ³ Decision Tree

**Definition:** A flowchart-like tree structure where each internal node represents a test on a feature, each branch a result, and each leaf a class label.

**Used for:** Classification & Regression tasks.

**Key Concepts:**
- Root Node, Decision Node, Leaf Node
- Impurity Measures: Gini Index, Entropy

**Pros:**

**Cons:**

## 4ðŸŒ² Random Forest

**Definition:** An ensemble of multiple Decision Trees trained on different subsets of data and features.

**Prediction:**
- Classification â†’ Majority Vote
- Regression â†’ Average of outputs

**Why it works better:**
- Reduces Overfitting
- More accurate and stable
- Handles missing values and high dimensionality

---

## ðŸ§  Overfitting

**Overfitting:** When the model learns noise from the training data and fails to generalize to new data.

- High accuracy on training data
- Poor performance on test data

**How to prevent:**
- Use Random Forest or ensemble methods
- Prune trees
- Use cross-validation


## 5ðŸ“Œ K-Nearest Neighbors (KNN)

### âœ… What is KNN?
KNN is a **supervised learning** algorithm used for both **classification** and **regression** (mostly classification).  
It classifies a new data point based on the **majority class** of its **K nearest neighbors** in the training data.

---

### ðŸ§  How It Works:
1. Choose the number **K** (e.g., 3, 5).
2. Measure the distance (commonly Euclidean) from the new point to all training points.
3. Select the **K nearest points**.
4. **Majority vote**: Assign the class with the most occurrences among the K neighbors.

---

### ðŸ“Š Real-life Analogy:
You move to a new city and want to know your likely favorite food. You ask your 5 nearest neighbors â€” if 3 like pizza and 2 like pasta, youâ€™re predicted to like **pizza**.

---

### ðŸ“ˆ Important Concepts:
- **Distance Metric**: Usually Euclidean
- **K Value**:
  - Too small â†’ Noisy, overfitting
  - Too large â†’ May include irrelevant neighbors
- **Lazy Learner**: KNN doesnâ€™t learn during training, it memorizes the data and computes during prediction.

---

### âœ… Advantages:
- Simple and easy to implement
- No training time
- Works well for small datasets

### âŒ Disadvantages:
- Slow on large datasets
- Sensitive to irrelevant features and scaling

---

### ðŸ“· Visualization Example:
- Each dot has a class label (ðŸ”µ or ðŸ”´).
- The unknown point (â­) is classified based on its nearest neighbors.



## 6ðŸŒ¼ Naive Bayes Algorithm (ML)

### ðŸ”¹ What is Naive Bayes?
Naive Bayes is a **supervised classification algorithm** based on **Bayes' Theorem**, assuming **independence between features**.

It's called "naive" because it assumes that all input features are **independent** of each other â€” which is rarely true in real data, but still works surprisingly well in practice.

---

### ðŸ”¹ Bayesâ€™ Theorem (Clean Format)

Bayes' Theorem is written as:

P(A | B) = [ P(B | A) * P(A) ] / P(B)

Where:
- **P(A | B)**: Posterior â†’ Probability of class A given feature B  
- **P(B | A)**: Likelihood â†’ Probability of feature B given class A  
- **P(A)**: Prior â†’ Probability of class A  
- **P(B)**: Evidence â†’ Probability of feature B


---

### ðŸ”¹ Types of Naive Bayes:
| Type            | Use Case                            |
|-----------------|-------------------------------------|
| GaussianNB      | Continuous features (e.g. iris data) |
| MultinomialNB   | Text classification (e.g. spam)     |
| BernoulliNB     | Binary/boolean features              |

---

### ðŸ”¹ When to Use:
- Text classification (spam detection, sentiment analysis)
- Medical diagnosis
- Real-time predictions

---

### âœ… Advantages:
- Simple and fast
- Works well on small datasets
- Handles high-dimensional data well (e.g. text)

### âš ï¸ Limitations:
- Assumes feature independence
- Performs poorly if features are correlated

---

### ðŸ§  Common Use Case:
- Spam Email Classifier using **Multinomial Naive Bayes**

---



````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train model
model = GaussianNB()
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
````

## 7ðŸ§  Support Vector Machine (SVM)

### What is SVM?
Support Vector Machine (SVM) is a supervised learning algorithm used primarily for classification. It finds the **optimal decision boundary (hyperplane)** that best separates data points of different classes by maximizing the margin between them.

---

### ðŸ”‘ Key Concepts

- **Hyperplane**: A line/plane that divides classes in the feature space.
- **Support Vectors**: Data points closest to the hyperplane that influence its position.
- **Margin**: Distance between the hyperplane and the support vectors. A larger margin generally leads to better generalization.

---

### ðŸ”„ Linearly vs Non-Linearly Separable Data

- **Linearly Separable**: A straight line or plane can separate the classes.
- **Non-Linearly Separable**: Requires mapping to a higher-dimensional space using a **kernel** function.

---

### ðŸ§ª Common SVM Kernels

| Kernel Type | Use Case |
|-------------|----------|
| **Linear** | When data is linearly separable |
| **Polynomial** | For curved boundaries |
| **RBF (Gaussian)** | For complex, non-linear problems |
| **Sigmoid** | Similar to neural network activation |

---

### âœ… Advantages
- Works well in high-dimensional spaces
- Effective with small- to medium-sized datasets
- Only depends on support vectors, making it memory efficient

### âš ï¸ Limitations
- Computationally expensive on large datasets
- Performance highly depends on kernel and hyperparameter tuning

---

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train SVM model with RBF kernel
model = SVC(kernel='rbf')
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluation
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
````
